```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore', DeprecationWarning)
import random

#import the data
df = pd.read_csv('Input/DefaultCreditcardClients.csv')
df.rename(columns={'default payment next month':'default'}, inplace=True)
df.index = df.ID
if 'ID' in df:
    del df['ID']
df.head()
```

```python
# Adding Random Numbers to the df.
np.random.seed(12)  #just keeps the same set of random numbers, so the results are stable through iterations.
df['RandNum'] = np.random.randint(0,10,df.shape[0])
#checking the distribution
pd.pivot_table(df,index = ['RandNum'], values = ['AGE'], aggfunc=[len])
```

```python
# Assigning each observation to test or Train based on Random values created above.
# I chose 3 and 4 because the added up to close to 6,000.
df['Set'] = ['Test' if x == 3 or x == 4 else 'Train' for x in df['RandNum']]
pd.pivot_table(df,index = ['Set'], values = ['AGE'], aggfunc=[len])

## Coldn't get code below to work:
#df['Train'] = np.where(df['RandNum']=3,'Test','Train')
#df['Set'] = pd.DataFrame.where('Test',(df['RandNum']==3 or df['RandNum']==4),'Train')
```
